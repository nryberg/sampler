# Coffee Sales

```python
import pandas as pd
```

# Source

```python
# Get Sales

df_sales = pd.read_csv('./Coffee_Sales.csv')
```

```python
df_sales.head()
```

```python
# Get Locations

df_location = pd.read_csv('./payees.csv')
df_location.head()
```

```python
# Mix together

df_sales_location = pd.merge(df_sales, df_location, on='Payee', how='inner')
df_sales_location.head()
```

```python
# Fix Data
# - First create a proper date field, then pull the year

df_sales_location['Date'] = pd.to_datetime(df_sales_location['Purchase_Date'])
df_sales_location['Year'] = df_sales_location['Date'].dt.year


df_sales_location.head()
```
